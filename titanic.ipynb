```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# TASK 1
print(df.shape)
print(df.columns.tolist())
print(df.isnull().sum())

# TASK 2
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df = df.drop(columns='Cabin')
print(df.isnull().sum())

# TASK 3
survived_count = df[df['Survived'] == 1].shape[0]   # FIXED
print(f"Survivors: {survived_count}")                # 342
print(f"Overall survival rate: {df['Survived'].mean():.2%}")  # 38.38%
print(df['Sex'].value_counts())

# TASK 4
print((df.groupby('Pclass')['Survived'].mean() * 100).round(2))
print(df.groupby('Pclass')['Fare'].mean().round(2))
print(df.groupby(['Sex', 'Pclass']).agg(
    SurvivalRate=("Survived", "mean")
).round(3))

# TASK 5
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)   # FIXED
survival_alone = df.groupby('IsAlone')['Survived'].mean().round(3)
print(survival_alone)
# IsAlone 0 (not alone): ~0.505
# IsAlone 1 (alone):     ~0.304

# TASK 6
top5 = df.sort_values('Fare', ascending=False)[   # FIXED
    ['Name', 'Fare', 'Pclass', 'Survived', 'Age']
].head(5)
print(top5)
```
